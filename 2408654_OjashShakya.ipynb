Task - 3 - Outlier Detection:

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.


```python
south_asian_df = whr_df[whr_df['Country name'].isin(south_asian_countries)]

q1_score = south_asian_df['score'].quantile(0.25)
q3_score = south_asian_df['score'].quantile(0.75)
iqr_score = q3_score - q1_score
print("Outliers based on their score:\n")
print("1st Quartile: ",q1_score)
print("3rd Quartile: ",q3_score)
print("Inter Quartile Range: ",iqr_score)

q1_gdp = south_asian_df['Log GDP per capita'].quantile(0.25)
q3_gdp = south_asian_df['Log GDP per capita'].quantile(0.75)
iqr_gdp = q3_gdp - q1_gdp
print("\nOutliers based on their GDP per capita:\n")
print("1st Quartile GDP: ",q1_gdp)
print("3rd Quartile GDP: ",q3_gdp)
print("Inter Quartile Range GDP: ",iqr_gdp)
```

2. Define outliers using the 1.5 × IQR rule.


```python
lower_bound_score = q1_score - 1.5 * iqr_score
upper_bound_score = q3_score + 1.5 * iqr_score
lower_bound_gdp = q1_gdp - 1.5 * iqr_gdp
upper_bound_gdp = q3_gdp + 1.5 * iqr_gdp
outlier_scores = south_asian_df[(south_asian_df['score'] < lower_bound_score) | (south_asian_df['score'] > upper_bound_score)]
outliers_gdp = south_asian_df[(south_asian_df['Log GDP per capita'] < lower_bound_gdp) | (south_asian_df['Log GDP per capita'] > upper_bound_gdp)]
outliers = pd.concat([outlier_scores, outliers_gdp])
print("Outliers in South Asia based on Score and GDP per Capita:")
outliers
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.

```python
plt.figure(figsize=(12, 6))
plt.scatter(south_asian_df['Log GDP per capita'], south_asian_df['score'], color='skyblue', label='south_asian_df Points')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='lightcoral', label='outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Outlier Detection in South Asia')
plt.grid(True, linestyle='--', alpha=0.8)
plt.legend()
plt.show()
```
